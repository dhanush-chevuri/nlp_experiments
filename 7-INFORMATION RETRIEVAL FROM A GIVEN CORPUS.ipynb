```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
corpus = [
    "The quick brown fox jumps over the lazy dog.",
    "Never jump over the lazy dog quickly.",
    "The dog is lazy but the fox is quick.",
    "A quick brown dog outpaces a quick fox.",
]
query = "quick fox"
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus + [query])
cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()
most_similar_doc_index = cosine_similarities.argmax()
print(f"Query: \"{query}\"")
print(f"Most similar document: \"{corpus[most_similar_doc_index]}\"")
print(f"Cosine Similarity: {cosine_similarities[most_similar_doc_index]:.4f}")
```
