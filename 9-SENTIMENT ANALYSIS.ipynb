```python

from transformers import pipeline

sentiment_analysis = pipeline("sentiment-analysis")
texts = [
    "I love working with natural language processing!",
    "This is a terrible day, nothing is going right.",
    "I am feeling quite neutral about the whole situation."
]
for text in texts:
    result = sentiment_analysis(text)
    label = result[0]['label']
    score = result[0]['score']
    print(f"Text: {text}\nSentiment: {label} (Confidence: {score:.2f})\n")
```
