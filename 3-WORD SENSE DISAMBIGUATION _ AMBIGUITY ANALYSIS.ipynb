```python
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
from nltk.wsd import lesk
import nltk

nltk.download('wordnet')
nltk.download('punkt')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

def disambiguate_word(sentence, target_word):
    """
    Disambiguate the meaning of a word in a sentence using Lesk's algorithm.

    :param sentence: The sentence containing the target word.
    :param target_word: The ambiguous word to disambiguate.
    :return: The most probable sense of the word and its definition.
    """

    if isinstance(sentence, list):
        sentence = ' '.join(sentence)

    tokens = word_tokenize(sentence)

    sense = lesk(tokens, target_word[0])

    if sense:
        return sense.name(), sense.definition()
    else:
        return None, "No suitable sense found."


sentence = ["I went to the bank to deposit money.",
            "The bank of the river was flooded after the storm"]

target_word = ["bank","bank"]

sense_name, sense_definition = disambiguate_word(sentence, target_word)
print(f"Word: {target_word}")
print(f"Sense: {sense_name}")
print(f"Definition: {sense_definition}")
```
